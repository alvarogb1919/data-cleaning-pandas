#### Voy a definir una función para la columna Year: Los valores son float y los voy a convertir a integers para que sea más limpio

```python
def integer_year(x):
    try:
        return int(x)
    except ValueError:
        return "Not available"
```

```python
integer_year(2022.0)
```

#### Aplico la funcion a la columna Year

```python
sharks_data["Year_clean"] = sharks_data["Year"].apply(integer_year)
```
